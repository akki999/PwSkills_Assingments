```python
import pandas as pd
import numpy as np
import seaborn as sns
from plotly import graph_objs as go
import plotly.express as px
```

```python
##Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
fig = px.scatter(x=df['age'], y=df['fare'])
fig.show()
```

```python
## Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
```

```python
df1 = sns.load_dataset('tips')
df1.head()
```

```python
fig = px.box(df1, y='total_bill')
fig.show()
```

```python
fig = px.box(df1, x="time", y="total_bill")
fig.show()
```

```python
## Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.
```

```python
df1.head()
```

```python
fig = px.histogram(df1, x='sex', y='total_bill', pattern_shape='smoker', color="day")
fig.show()
```

```python
## Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.
```

```python
df2 = sns.load_dataset('iris')
df2.head()
```

```python
fig = px.scatter_matrix(df2, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"], color="species")
fig.show()
```

```python
## Q5. What is Distplot? Using Plotly express, plot a distplot.
```

```python
## A Distplot or distribution plot, depicts the variation in the data distribution. 
## Distplot represents the overall distribution of continuous data variables. 
## It is used to depict the distplot with different variations in it.
```

```python
from plotly.figure_factory import create_distplot

df = pd.DataFrame({'2012': np.random.randn(200),
                   '2013': np.random.randn(200)+1})
fig = create_distplot([df[c] for c in df.columns], df.columns)
fig.show()
```
