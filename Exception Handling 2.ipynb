Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans: 
    Lookup Error acts as a base class for the exceptions that occur when a key or index used on a mapping or sequence of a list/dictionary is invalid or does not exists.

    The two types of exceptions raised are:

    a. IndexError
    b. KeyError

    Key Error:
        If a key you are trying to access is not found in the dictionary, a key error exception is raised.
    
    Index Error:
        When you are trying to access an index (sequence) of a list that does not exist in that list or is out of range of that list, an index error is raised.

```python
## Key error:
try:  
    a = {1:'a', 2:'b', 3:'c'}  
    print (a[4])  
except LookupError:  
    print ("Key Error Exception Raised.")
else:  
    print ("Success, no error!")
```

```python
## Index Error:
try:  
    a = ['a', 'b', 'c']  
    print (a[4])  
except LookupError:  
    print ("Index Error Exception Raised, list index out of range")
else:  
    print ("Success, no error!")
```

Q5. Explain ImportError. What is ModuleNotFoundError?

Import Error: 
    
    This error generally occurs when a class cannot be imported due to one of the following reasons:

    a. The imported class is in a circular dependency.
    b. The imported class is unavailable or was not created.
    c. The imported class name is misspelled.
    d. The imported class from a module is misplaced.
    e. The imported class is unavailable in the Python library.
    
Module Not Found error:
    
    As the name implies, this error occurs when you're trying to access or use a module that cannot be found. In the case of the title, the "module named Python" cannot be found.
    
    Here are a few reasons why a module may not be found:

    a. you do not have the module you tried importing installed on your computer
    b. you spelled a module incorrectly (which still links back to the previous point, that the misspelled module is not installed)...for example, spelling numpy as numpys during import
    c. you use an incorrect casing for a module (which still links back to the first point)...for example, spelling numpy as NumPy during import will throw the module not found error as both modules are "not the same"
    d. you are importing a module using the wrong path
